```python
import json
from pprint import pprint

with open("outputs/reports/multipartite_003e/cheat_eval/cheat_clusters/summary.json", "r") as f:
    cheat = json.load(f)
print(cheat.keys())
```

```python
ccs = {}
for cluster in cheat["layer_2_top_k_14_0"]["clusters"]:
    ccs_t = {}
    ccs_t ["component"] = cluster["cluster_label"]
    if ccs_t["component"] == "noise":
        ccs_t["r2"] = None
    else:
        ccs_t["r2"] = cluster["cluster_component_r2"][ccs_t["component"]]["mean_r2"]
    ccs_t["all_r2"] = {}
    for k in cluster["cluster_component_r2"]:
        ccs_t["all_r2"]["r2_" +k] = cluster["cluster_component_r2"][k]["mean_r2"]
    ccs_t["geometry"] = cluster["geometry"]["best_geometry"]
    ccs_t["gw_dist"] = cluster["geometry"]["best_optimal_distance"]
    ccs_t["latent_count"] = cluster["latent_count"]
    ccs[ccs_t["component"]] = ccs_t
    if "mess" in ccs_t["component"]:
        ccs_t["true_geometry"] = "simplex_2"
        ccs_t["true_geometry_distance"] = cluster["geometry"]["all_fits"]["simplex_2"]["optimal_distance"]
    elif "quantum" in ccs_t["component"]:
        ccs_t["true_geometry"] = "circle"
        ccs_t["true_geometry_distance"] = cluster["geometry"]["all_fits"]["circle"]["optimal_distance"]
```
